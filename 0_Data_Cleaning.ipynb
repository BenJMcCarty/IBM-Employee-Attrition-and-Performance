# Exploring the Data

# Importing Packages and Reading Data

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
path = './data/WA_Fn-UseC_-HR-Employee-Attrition.csv'
```

```python
data = pd.read_csv(path)
data
```

# Starting EDA

```python
data.info()
```

## Numeric Features

```python
numeric_df = data.select_dtypes(include = "number")
numeric_df
```

```python
numeric_df.describe(include = "number").T
```

```python
## Inspect total unique values per feature to identify any categorical/ordinal features

numeric_df.nunique().sort_values()
```

## Drop Constant Features

```python
constant_feats = (numeric_df
                  .nunique()
                  [numeric_df.nunique() == 1]
                  .index)
constant_feats
```

```python
numeric_df = numeric_df.drop(columns=constant_feats)
numeric_df
```

## Investigate Possible Categorical Numeric Features

```python
## Slice out feature names with total unique values between 1 and 5

cat_num_cols = (numeric_df.nunique()                                        ## Calculate number of unique features
                .sort_values()                                              ## Sort by ascending values
                [(numeric_df.nunique() <= 5) & (numeric_df.nunique() > 1)]  ## Total unique features between 1 and 5
                .index)                                                 ## Convert to list
cat_num_cols
```

```python
## Review values for categorical numeric features
numeric_df.loc[:, cat_num_cols]
```

```python
## Loop through categorical numeric features to print unique values in each feature

for col in cat_num_cols:
    print(col, ": \t", sorted(numeric_df.loc[:, col].unique()))
```

## Split Numeric Data b/t Categorical & Continuous

### Categorical DataFrame

```python
## Slice out categorical features from numeric dataframe

cat_num_df = numeric_df.loc[:, cat_num_cols]
cat_num_df
```

```python
cat_num_df.describe().T
```

#### Save Results

```python
cat_num_df.to_excel('./data/data_subset_ordinal.xlsx', index = False)
```

### Continuous DataFrame

```python
## Drop categorical features and EmployeeNumber from numeric dataframe

cont_num_df = numeric_df.drop(columns=cat_num_cols)
cont_num_df
```

```python
cont_num_df.describe().T
```

#### Save Results

```python
cont_num_df.to_excel('./data/data_subset_continous.xlsx', index = False)
```
