## Numeric Features

```python
data.describe(include = "number").T
```

```python
numeric_df = data.select_dtypes(include = "number")
numeric_df
```

```python
numeric_df.nunique().sort_values()
```

```python
cat_num_cols = numeric_df.nunique().sort_values()[numeric_df.nunique() <= 5].index
cat_num_cols
```

```python
numeric_df.loc[:, cat_num_cols]
```

```python
for col in cat_num_cols:
    print(col, ": \t", numeric_df.loc[:, col].unique())
```

```python
## Slice out categorical features from numeric dataframe

cat_num_df = numeric_df.loc[:, cat_num_cols]
cat_num_df
```

```python
## Drop categorical features from numeric dataframe

cont_num_df = numeric_df.drop(columns=cat_num_cols)
cont_num_df
```
